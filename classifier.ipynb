```python
import pandas as pd
import re
```

```python
constructed_dataset = pd.read_csv("twitterdataset.csv")
constructed_dataset['labels(0:non-scam; 1:scam)'] = constructed_dataset['labels(0:non-scam; 1:scam)'].fillna(0)
constructed_dataset['labels(0:non-scam; 1:scam)'] = constructed_dataset.apply(lambda row: bool(row['labels(0:non-scam; 1:scam)']), axis=1)
constructed_dataset['tweet'] = constructed_dataset[['tweet']].astype('str')
constructed_dataset
```

```python
constructed_dataset.to_csv('twitterdataset.csv')
```

```python
def classifier(string):
    contains_btc_add = bool(re.search("[13][a-km-zA-HJ-NP-Z1-9]{25,34}", string))
    contains_eth_add = bool(re.search("0x[a-fA-F0-9]{40}$", string))
    phrases = ['legit', 'legitimate', 'send me', 'double', 'whatsapp']
    legit_bot_phrases = ['transferred from', 'move from']
    contains_scam_phrase = False
    for phrase in phrases:
        if phrase in string.lower():
            contains_scam_phrase = True
    contains_legit_phrase = False
    for phrase in legit_bot_phrases:
        if phrase in string.lower():
            contains_legit_phrase = True
    return ((not contains_legit_phrase) and (contains_btc_add or contains_eth_add or contains_scam_phrase))
```

```python
constructed_dataset['classifier_labels'] = constructed_dataset.apply(lambda row: classifier(row['tweet']), axis=1)
```
